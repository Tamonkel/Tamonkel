```python
#a
def isReflexive(M, R): #M menge # Tuple
    for i in M: #jedes element in M durch
        if (i, i) not in R: #schaut ob i in relation mit sich selbst steht
            return False
    return True
```

```python
isReflexive ([1 ,2 ,3] ,[(1 ,2) ,(1 ,1) ,(2 ,2) ,(2 ,1) ,(3 ,3)]) 
#(4,3) false
```

```python
isReflexive ([1 ,2 ,3] ,[(1 ,1) ,(2 ,2)])
```

```python
#b
def isSymmetric(M, R):
    for a in M:
        for b in M:
            if (b, a) not in R: #wenn a,b auch b,a esxestiert sonst kein tupel
                return False
    return True
```

```python
isSymmetric ([1 ,2 ,3] ,[(1 ,2) ,(2 ,1) ,(2 ,2) ,(1 ,3) ,(3 ,1)])
```

```python
isSymmetric ([1 ,2 ,3] ,[(1 ,2) ,(2 ,1) ,(2 ,2) ,(1 ,3)])
```
