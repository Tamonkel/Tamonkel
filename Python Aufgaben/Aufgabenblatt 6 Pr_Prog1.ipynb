```python
class stack:
    def __init__(self):
        self.st = []

    #a
    def push(self, element):
        self.st.append(element) #append fügt zur liste hinzu

    #b    
    def pop(self):
        return self.st.pop() #pop entfernt oberstes element von stack

    #c    
    def __len__(self):
        return len(self.st) # gibt die anzahl elemente aus
    
    #d
    def toList(self):
        return self.st.copy() #gibt stack zurück
    
    #e
    def multPop(self, n):
        ergebnis = []
        for i in range(n): #n mal durchlaufen
            ergebnis.append(self.pop())  
        return ergebnis     #Ergebnisse der Pop-Operation als liste zurückliefern
```

# Aufgabe 6.1

```python
class Stack:
    def __init__(self):
        self.st = []

    def push(self, element):
        self.st.append(element) #append fügt zur sack hinzu

    def pop(self):
        if self.st:
            return self.st.pop() #pop entfernt oberstes element von stack<s
        return None

    def __len__(self):
        return len(self.st)  # gibt die anzahl elemente aus

    def toList(self):
        return self.st.copy() #gibt stack zurück

    def multPop(self, n):
            ergebnis = []
            for i in range(n):
                ergebnis.append(self.pop())   
            return ergebnis #Ergebnisse der Pop-Operation als liste zurückliefern

class AufgabenManager:
    def __init__(self):
        self.aufgaben = {}

    # a Neue Aufgabe mit bestimmter Priorität hinzufügen
    def neueAufgabe(self, aufgabe, prio):
        if prio not in self.aufgaben:
            self.aufgaben[prio] = Stack()
        self.aufgaben[prio].push(aufgabe)

    #b Höchste (kleinste) Prioritätsstufe zurückliefern
    def hoechstePrio(self):
        if not self.aufgaben:
            return None
        return min(p for p in self.aufgaben if len(self.aufgaben[p]) > 0)

    #c Nächste Aufgabe mit höchster Prio erledigen (aus der Datenstruktur loescht und den Aufgaben-String zuruckliefert)
    def erledigeNaechsteAufgabe(self):
        prio = self.hoechstePrio()
        if prio is None:
            return None
        return self.aufgaben[prio].pop() #wenn leer soll b entfernt weren!!!

    #d Alle Aufgaben mit bestimmter Priorität zurückgeben
    def alleAufgabenMitPrio(self, prio):
        if prio in self.aufgaben:
            return self.aufgaben[prio].toList()
        return []

    #e  Liste aller Prioritaetsstufen zuruckliefern
    def allePrios(self):
        return list(self.aufgaben.keys())

    #f Anzahl Aufgaben mit bestimmter Priorität soll eine Prioritaetsstufe ubergeben bekommenn
    def anzahlAufgabenPrio(self, prio):
        if prio in self.aufgaben:
            return len(self.aufgaben[prio])
        return 0

    #g Anzahl der Aufgaben
    def anzahlAufgaben(self):
        return sum(len(stack) for stack in self.aufgaben.values())
```

```python
aufgs = AufgabenManager()
aufgs.neueAufgabe("Küche putzen", 5)
aufgs.neueAufgabe("Auf Prog 1 lernen", 1)
aufgs.neueAufgabe("Oma besuchen", 2)
aufgs.neueAufgabe("Auf Mathe 1 lernen", 1)
aufgs.neueAufgabe("Fahrrad putzen", 10)

print(aufgs.erledigeNaechsteAufgabe())
print(aufgs.erledigeNaechsteAufgabe())
print(aufgs.hoechstePrio())
print(aufgs.anzahlAufgabenPrio(1))
print(aufgs.anzahlAufgaben())
```
