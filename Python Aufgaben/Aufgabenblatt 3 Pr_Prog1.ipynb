```python
import math
import random 

def scheresteinpapier():
    Spielzüge = ["Schere","Stein","Papier"] #mögliche Spielzüge

    #Zug Computer = zufälliger string aus Liste "Spielzüge"
    Zug_Computer = Spielzüge[random.randint(0,2)] 

    #Zug des Spielers
    print(Zug_Computer)
    Zug_Spieler = input("wähle einen Zug")

    #Wenn Gleichstand
    if str(Zug_Computer) == str(Zug_Spieler):
        print("Gleichstand spiele nochmal")
        return scheresteinpapier()#Funktion nochmal aufrufen
    
    #Eingabe Überprüfen
    if Zug_Spieler not in Spielzüge:
        print("Falsche Eingabe")
        return 

    #Gewinn bedingungen 
    if Zug_Spieler == "Schere" and Zug_Computer == "Papier": #Schere schlägt Papier
        print("Gewonnen")
    elif Zug_Spieler == "Stein" and Zug_Computer == "Schere": #Stein schlägt Schere
        print("Gewonnen")
    elif Zug_Spieler == "Papier" and Zug_Computer == "Stein": #Papier schlägt Stein
        print("Gewonnen")
    else:
        print("Verloren Spiele nochmal")        
        return scheresteinpapier()#Funktion nochmal aufrufen

scheresteinpapier()
```

```python
import random

def ratespiel(n):
    zielzahl = random.randint(0, n)
    versuche = 0  
    
    while True:
        try:
            raten = int(input("Rate eine Zahl: ")) #zahl eingeben
            versuche += 1  
            
            if raten > zielzahl: #prüft ob die zahl zu groß ist
                print("zu groß")
            elif raten < zielzahl: #prüft ob diezahl zu klein ist
                print("zu klein")
            else:
                print("richtig") #richtig geraten
                return versuche  
        except ValueError: #Fehler bei falscher eingabe 
            print("Bitte eine gültige Ganzzahl eingeben.")

ratespiel(10) #(100)
```

```python
from itertools import product #

def vier_buchstaben_worte():
    vokale = {'a', 'e', 'i', 'o', 'u'}  # Vokale
    alphabet = 'abcdefghijklmnopqrstuvwxyz'  # Alphabet

    ergebnis = []  # Liste für gefilterte Wörter

    for buchstaben in product(alphabet, repeat=4):  # Alle möglichen Kombinationen
        wort = ''.join(buchstaben)
        if wort[1] not in vokale and wort[2] not in vokale:
            ergebnis.append(wort)  # Füge Wort hinzu, wenn Bedingung erfüllt ist

    return ergebnis

alle_worte = vier_buchstaben_worte() #speichert wörter in alle_worte
print(f"Anzahl der gültigen Wörter: {len(alle_worte)}")  #anzahl gültiger worte
print(alle_worte[:10])  # Zeige die ersten 10 gültigen Wörter
```

```python
def vier_buchstaben_worte():
    vokale = 'a', 'e', 'i', 'o', 'u' #vokale
    alphabet = 'abcdefghijklmnopqrstuvwxyz' #alphatbet
    
    worte = ''.join(buchstaben) for buchstaben #alle möglichen kombis aus 4 buchstaben
    
    ergebnis = [] #liste gefilterte wörter
    for wort in worte: 
        #prüft ob zweite und dritte buchstaben keine vokale sind
        if not (wort[1] not in vokale and wort[2] not in vokale):
            ergebnis.append(wort) #füg zu ergebnis hinzu
    
    return ergebnis


alle_worte = vier_buchstaben_worte() #speichert wörter in alle_worte
print(f"Anzahl der gültigen Wörter: {len(alle_worte)}") #anzahl gültiger worte
print(alle_worte[:10])  #gibt die 10 ersten gültigen wörter aus
```

```python
def vier_buchstaben_worte():
    vokale = {'a', 'e', 'i', 'o', 'u'}
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    ergebnis = []

    for buchstabe1 in alphabet:
        for buchstabe2 in alphabet:
            for buchstabe3 in alphabet:
                for buchstabe4 in alphabet:
                    wort = buchstabe1 + buchstabe2 + buchstabe3 + buchstabe4
                    if buchstabe2 not in vokale and buchstabe3 not in vokale:
                        ergebnis.append(wort)

    return ergebnis

alle_worte = vier_buchstaben_worte()
print(f"Anzahl der gültigen Wörter: {len(alle_worte)}")
print(alle_worte[:10])  # Zeigt die ersten 10 gültigen Wörter
```
