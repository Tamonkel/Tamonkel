# Aufgabe 5.0

```python


import json
import os

def saveperson(n, v, a, g):

    # Überprüfen ob es di richtigen Eingabetypen sind
    assert isinstance(n, str)
    assert isinstance(v, str)
    assert isinstance(a, int)
    assert isinstance(g, int) 

    person = {
        "Nachname": n,
        "Vorname": v,
        "Alter": a,
        "Groesse": g,
    }#dictionary

    file_name = "namensliste.json"

    # JSON-Datei laden odr leere Liste erstellen
    if os.path.exists(file_name):
        with open(file_name, "r") as file: #read
            personen = json.load(file)
    else:
        personen = []

    personen.append(person) #fügt person zu liste hinzu


    with open('namensliste.json', 'w') as f: #schreibt aktualisierte liste in die json datei
        json.dump(personen, f) #speichert liste
                            
    print(f"Person {n} {v} wurde gespeichert.")
```

```python
saveperson("Meier", "Tom", 25, 170)
```

```python
#a
def finderperson(n):
   assert isinstance(n, str) # Der Name muss ein String sein

   filename = "namensliste.json"

   with open(filename, "r") as file: # file öffnen als file
       data = json.load(file)  

   for person in data:        
       if person ["Nachname"] == n:
           return person # Gesamte information ausgeben

   return "Person nicht gefunden."
```

```python
person = finderperson("Meier")
print(person)
```

```python
import json
import os

def removeperson(n):
    # Überprüfen, ob der Name ein String ist
    assert isinstance(n, str)  # Der Name muss ein String sein

    filename = "namensliste.json"

    # Laden der Daten aus der JSON-Datei
    with open(filename, "r") as file:
        data = json.load(file)

    #Person mit dem angegebenen Nachnamen entfernen
    gefilterte_daten = [person for person in data if person["Nachname"] != n]

    #Person mit diesem Nachnamen gefunden?
    if len(gefilterte_daten) == len(data):
        return "Person nicht gefunden."

    # Die gefilterten Daten in der JSON-Datei speichern
    with open(filename, "w") as file:
        json.dump(gefilterte_daten, file) #speichern

    return "Person erfolgreich entfernt."
```

```python
result = removeperson("Meier")
print(result)
```

```python
person = finderperson("Meier")
print(person)
```

```python
import os
import re

def findFile(d, n):
    if not os.path.exists(d): # Überprüfen, ob das Verzeichnis existiert
        raise Exception("Das Verzeichnis existiert nicht.")

    dateien = os.listdir(d) # Listet die Dateien im Verzeichnis auf

    gefundene_dateien = [datei for datei in dateien if re.fullmatch(n, datei)] # Dateien werden gefiltert, die dem Muster entsprechen

    if gefundene_dateien: # Ergebnisse ausgeben
        print("Gefundene Dateien:")
        for datei in gefundene_dateien:
            print(datei)
    else:
        print("Keine Dateien gefunden")
```

# Aufgabe 5.2

```python
class MyCompl2:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def __str__(self):
        return f"{self.real} + {self.imag}i" # darstellung zahl als zeichenkette

    def __repr__(self):
        return f"({self.real}, {self.imag})i" # .. als interpreter

    def __add__(self, other):
        return MyCompl2(self.real + other.real, self.imag + other.imag) # addition kompl zahlen

    def __mul__(self, other):
        real = self.real * other.real - self.imag * other.imag  #multiplikation kompl zahlen -
        imag = self.real * other.imag + self.imag * other.real #+
        return MyCompl2(real, imag)
```

```python
#a und b.)
c1 = MyCompl2 (1.0 ,2.0)
```
