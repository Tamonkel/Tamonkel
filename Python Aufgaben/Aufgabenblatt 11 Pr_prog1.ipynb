```python
#a.) Sucht nach Dateien mit `.tex` und 'aufg' im Namen

files_with_tex_aufg = [
    os.path.join(root, file) #fügt daei mit vollständigem pfad zur liste hinzu
    for root, i, files in os.walk('.') #durchsucht verzeichnisse im aktuellen ordner
    for file in files
    if file.endswith('.tex') and 'aufg' in file #schaut ob datei mit .tex endet und aufg in namen hat
]
print(files_with_tex_aufg)
```

```python
list(os.walk('.'))
```

```python
#b Sucht nach .tex -Dateien mit über 100 kurzen Zeilen

kurzezeilen = [
    os.path.join(root, file) #fügt daei mit vollständigem pfad zur liste hinzu
    for root, i, files in os.walk('.') #durchsucht verzeichnisse im aktuellen ordner
    for file in files
    if file.endswith('.tex') and sum(len(line.strip()) < 2 for line in open(os.path.join(root, file))) > 100 #überprüft zeilenlänge
]
print(kurzezeilen)
```

```python
#c Sucht nach .tex -Dateien mit einer Zeile, die mehr als 30 Wörter enthalten

langezeilen = [
    os.path.join(root, file)
    for root, i, files in os.walk('.')
    for file in files
    if file.endswith('.tex') and any(len(line.split()) > 30 for line in open(os.path.join(root, file))) #dateei die mehr als 30 wörter enthält
]              #datei die mit .text endet 
print(langezeilen)
```

```python
#b Array von 0 bis 1000 mit Zahlen, nicht durch 11 und 13 teilbar sind
zahlen = np.arange(1001)
filtern = zahlen [(zahlen % 11 != 0) & (zahlen % 13 != 0)]

print(filtern)
```

```python
#c Array Länge 100 mit Nullen an geraden, Einsen an ungeraden Indexpositionen
muster = np.zeros(100, dtype=int)

muster[1::2] = 1 # setzt an allen ungeraden stellen eine 1 jede 2te stelle

print(muster)
```
