# Aufgabe 10

```python
#a
from functools import reduce

Liste = [(1,10), ('a','b'), ([1], [2])]

einfacheliste = reduce(lambda x, y: x + list(y), Liste, []) #x,y: operation , x + list(y) wandelt jedes tupel in eine liste um und fügt x hinzu

print(einfacheliste)
```

```python
#b

Liste = [3, 6, 12, 1, 20]

def max(lst):
    return reduce(lambda x, y: x if x > y else y, lst) #wenn x  größer sonst soll y verwendet werden

print(max(Liste))
```

```python
#a sich die Liste der L¨angen der Zeilen der Datei test.txt ausgeben zu lassen.

[len(zeile) for zeile in open("test.txt")] #länge der zeilen
```

```python
#b die ersten drei Zeilen der Datei test.txt ausgeben zu lassen.

[zeile.strip() for zeile in open("test.txt")][:3] #ersten drei zeilen ausgeben
```

```python
#c die erste Zeile der Datei test.txt in rückwärts ausgeben zu lassen.

open("test.txt").readline().strip()[::-1] # erste Zeile rückwärts ausgeben
```

```python
#d ) sich die Liste der L¨angen des ersten Wortes jeder Zeile aus test.txt ausgeben zu lassen

[len(line.split()[0]) if line.split() else 0 for line in open("test.txt")] # Längen des ersten Wortes jeder Zeile ausgeben
```

```python
#e sich alle Zeilennummern der Datei test.txt ausgeben zu lassen, die keine Leerzeichen enthalten.

[i for i, line in enumerate(open("test.txt"), start=1) if ' ' not in line.strip()] # Zeilennummern finden, die keine Leerzeichen enthalten
```

```python
#f  sich alle Dateinamen des aktuellen Verzeichnisses ausgeben zu lassen, deren Namen mit .py enden.

[f for f in os.listdir('.') if f.endswith('.py')]# Dateien im Verzeichnis, die mit .py enden
```

```python
#g sich alle Dateien des aktuellen Verzeichnisses ausgeben zu lassen, deren Namen mehr als 10
#Zeichen lang ist und deren Inhalt mehr als 10 Zeilen lang ist.


[f for f in os.listdir('.') if len(f) > 10 and len(open(f).readlines()) > 10] # Filtert die Dateien aus
```
