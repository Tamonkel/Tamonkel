# Aufgabe 9

```python
import random
#a 
zufall = [random.randint(1, 6) for i in range(10)] #10 zufällige Werte zwischen 1 und 6
print(zufall)
```

```python
#b Wahrscheinlichkeit, dass bei 10 Würfen keine 1 fällt
simulationen = 100000
keine_eins = sum(
    all(random.randint(1, 6) != 1 for i in range(10)) for i in range(simulationen)
)
wahrscheinlichkeit = keine_eins / simulationen
print(f"Wahrscheinlichkeit: {wahrscheinlichkeit}")
```

```python
#c Wahrscheinlichkeit, dass genau 3-mal eine 1 fällt
simulationen = 100000
drei_einsen = sum(
    sum(random.randint(1, 6) == 1 for i in range(10)) == 3 for i in range(simulationen)
)
wahrscheinlichkeit = drei_einsen / simulationen
print(f"Wahrscheinlichkeit: {wahrscheinlichkeit}")

```

```python
import matplotlib.pyplot as plt

#a Funktion, die alle Teiler einer Zahl findet
def teiler(n):
    return [i for i in range(1, n + 1) if n % i == 0]
```

```python
teiler(45)
```

```python
#b Zahlen zwischen 1 und 1000 mit genau 8 Teilern
teiler = [n for n in range(1, 1001) if len(teiler(n)) == 8]
print(teiler)
```

```python
def teiler(n):
    return [i for i in range(1, n + 1) if n % i == 0]
```

```python
#c Zahl zwischen 1 und 1000 mit den meisten Teilern
meisteteiler = max(range(1, 1001), key=lambda x: len(teiler(x)))
print("meiste Teiler:", meisteteiler)
```

```python
#(d) Plot der Anzahl der Teiler für Zahlen von 1 bis 10000
import matplotlib.pyplot as plt
x_werte = range(1, 10001) ####
y_werte = [len(teiler(x)) for x in x_werte]  # teiler(x)  teiler x als liste und len() anzahhl der teiler

plt.figure(figsize=(10, 6))
plt.plot(x_werte, y_werte, color='blue', linewidth=0.5)
plt.grid(True)
plt.show()
```
