# Aufgabe 2.0

```python
def drei_zahlen(x,y,z):
    if (x > z and y > z): #wenn x und y größer z print summe aus x und y
        summe1 = (x + y)  
        return(summe1)
        #print("Fall 1") #kontrolle

    elif (x > y and z > y): #wenn x und z größer z print summe aus x und z
        Summe2 = (x + z)
        return(Summe2)
        #print("Fall 2")

    else:
        summe3 = (y + z) #wenn keines der oberen zutrift y + z
        return(summe3)
        #print("Fall 3")

drei_zahlen(1,2,3)
```

# Aufgabe 2.1

```python
def palindrom(wort):
    wort = wort.lower() #alle buchstaben kleinmachen

    if wort == wort [:: -1]: #wort rückswärts gelesen
        return True
    else: 
        return False

print(palindrom('Lagerregal'))

```

# Aufgabe 2.2

```python
def maxword(wort):
    text = wort.split()
    längste = max(text, key = len)
    return längste

maxword("was ist das längste wort")
```

# Aufgabe 2.4

```python
def anagram(wort1, wort2):
    wort1 = wort1.lower()
    wort2 = wort2.lower()
    
    if (sorted(wort1) == sorted(wort2)):
        return True
    else:
        return False

anagram('Erbgut', 'Betrug')
```
