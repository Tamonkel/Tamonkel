# Aufgabe 4.0

```python
def temperature(s):
 
    if not isinstance(s, str): #prüft ob es ein String ist
        raise ValueError("Eingabe muss ein String sein.")

    try:
        value = float(s[:-1])  #zahl überprüfen             vordere zeichen auser das letzte
        einheit = s[-1].upper()  #buchstaben überprüfen     (letztes zeichen)
    except ValueError:
        raise ValueError("Falsches Format") #wenn falsches foirmat fehler ausgebebn

    
    if einheit not in ('C', 'F', 'K'):
        raise ValueError("Die Einheit muss 'C' 'F' oder 'K' sein.") #richtige einheit?

    # umrechnungen der einheiten
    if einheit == 'C':
        celsius = value
        fahrenheit = celsius * 9 / 5 + 32
        kelvin = celsius + 273.15
    elif einheit == 'F':
        fahrenheit = value
        celsius = (fahrenheit - 32) * 5 / 9
        kelvin = celsius + 273.15
    elif einheit == 'K':
        kelvin = value
        celsius = kelvin - 273.15
        fahrenheit = celsius * 9 / 5 + 32

    return { #rückgabe der werte 
        'K': round(kelvin, 2), 
        'C': round(celsius, 2),
        'F': round(fahrenheit, 2)
    }
```

```python
erg = temperature("0C") # 0 Celsius
print(erg)
erg = temperature("0F") # 0 Fahrenheit
print(erg)
erg = temperature("0K") # 0 Kelvin
print(erg)
```

# Aufgabe 4.2

```python

def Primtest(n):
    if not isinstance(n, int): #prüft ob es eint Integer ist
     raise TypeError("kein Integer")
    
    for i in range(2, n): #prüft zahlen ob durch i teilbar ist
       if n % i == 0: 
          return False #wenn 0 dann keine primzahl
    else:
       return True

def zwillingsprimzahlen(n):
    if not isinstance(n, int):  #prüft ob es eint Integer ist
        raise TypeError("Kein Integer")
    
    zwillingsprimzahlen = [] #leere Liste
    for p in range(2, n - 1): #durchläuft alle zahlen von 2 bis  n-2
        if Primtest(p) and Primtest(p + 2): #sind beide Primzahlen?
            zwillingsprimzahlen.append((p, p + 2)) #zur Liste hinzufügen

    
    return zwillingsprimzahlen

zwillingsprimzahlen(10000)
```
